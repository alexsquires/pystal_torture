```python
import matplotlib.pyplot as plt
```

```python
fig,ax = plt.subplots(nrows=1, ncols=1, figsize=(10,10))
ax.set_xlim((0,51))
ax.set_ylim((0,50))
plt.axis('off')

for x in range(1,50,5):
    
    for y in range(1,50,5):
        
        
        circle1 = plt.Circle((x,y), 0.5 , color='b',fill=False)
        circle2 = plt.Circle((x+2.5,y-2.5),0.5,color= 'blue',fill=False)
        circle3 = plt.Circle((x+2.5,y),0.75,color='green',fill=False)
        ax.add_artist(circle1)
        ax.add_artist(circle2)
        ax.add_artist(circle3)
        if y==26:        
            plt.arrow(x,y,1.,-1.,head_width=1.5,head_length=1.5)
            plt.arrow(x+2.5,y-2.5,1,1,head_width=1.5,head_length=1.5)
```
